```python
import pandas as pd 
df1 = pd.read_csv("firstjoin.csv",index_col='date', parse_dates=['date'])
df1.head()
df1.dtypes
```

```python
df2 = pd.read_csv("activeadd.csv", parse_dates=['date'])
df2['date'] = df2['date'].dt.strftime('%Y-%m-%d')
df2["date"] = df2["date"].astype("datetime64[ns]")
df2["address"] = df2["address"].astype("float")
df2 = df2.set_index('date')
df2.dtypes
```

```python
df3 = pd.read_csv("mined.csv", parse_dates=['date'])
df3['date'] = df3['date'].dt.strftime('%Y-%m-%d')
df3["date"] = df3["date"].astype("datetime64[ns]")
df3["mined"] = df3["mined"].astype("float")
df3 = df3.set_index('date')
df3
```

```python
result = pd.merge(df1, df2, left_index=True, right_index=True)
result = pd.merge(result, df3, left_index=True, right_index=True)
result.to_csv('btcjoin.csv')

result
```

```python
result = result.drop(columns=['volume','change', 'low', 'high', 'open'])
result = result.rename(columns={"value": "wallets"})
result['price']=result['price'].str.replace(',','')
result['price']=result['price'].astype("float")
result['Metcafe']=result['address']**2
# df["Metcafe"] = df["Metcafe"].map("{:.1f}".format)
result['value'] = result['Metcafe']/result['mined']
result["value"] = result["value"].map("{:.2f}".format)
result['value']=result['value'].astype("float")
result['networkvalue'] = result["price"] - result["value"]

result
```
